```python
import pandas as pd
import tensorflow_data_validation as tfdv
import os


BASE_INPUT = './data/'


train_df = pd.read_csv(os.path.join(BASE_INPUT,'train.csv'))
val_df = pd.read_csv(os.path.join(BASE_INPUT,'val.csv'))
test_df = pd.read_csv(os.path.join(BASE_INPUT,'test.csv'))


```

```python
train_stats = tfdv.generate_statistics_from_dataframe(train_df)
```

```python
tfdv.visualize_statistics(train_stats)
```

```python
schema = tfdv.infer_schema(statistics=train_stats)
tfdv.display_schema(schema=schema)
```

```python
eval_stats = tfdv.generate_statistics_from_dataframe(val_df)
test_stats = tfdv.generate_statistics_from_dataframe(test_df)

# Compare evaluation data with training data
tfdv.visualize_statistics(lhs_statistics=eval_stats, rhs_statistics=train_stats,
                          lhs_name='EVAL_DATASET', rhs_name='TRAIN_DATASET')
```

```python
tfdv.visualize_statistics(lhs_statistics=test_stats, rhs_statistics=train_stats,
                          lhs_name='TEST_DATASET', rhs_name='TRAIN_DATASET')
```

```python
anomalies = tfdv.validate_statistics(statistics=eval_stats, schema=schema)
tfdv.display_anomalies(anomalies)
```

```python
from tensorflow.python.lib.io import file_io
from google.protobuf import text_format

file_io.recursive_create_dir(BASE_INPUT)
schema_file = os.path.join(BASE_INPUT, 'schema.pbtxt')
tfdv.write_schema_text(schema, schema_file)
```
